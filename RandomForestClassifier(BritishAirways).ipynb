```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
```

```python
y = df_1['YES']

X= df_1.drop('YES', axis = 1)

X_train,X_test,y_train,y_test = train_test_split(X,y, stratify = y, test_size = 0.25, random_state = 0)
```

```python
rf = RandomForestClassifier(random_state = 0)
```

```python
cv_params = {'n_estimators' : [50,300],
              'max_depth' : [10,50],
              'min_samples_leaf' : [0.5,1],
              'min_samples_split' : [0.001, 0.01],
              'max_features' : ["sqrt"],
              'max_samples' : [.5,.9]}
```

```python
scoring = {'accuracy', 'precision', 'recall', 'f1'}
```

```python
rf_cv = GridSearchCV(rf, cv_params, scoring=scoring, cv=5, refit='f1')
```

```python
%%time
rf_cv.fit(X_train, y_train)
```

```python
rf_cv.best_params_
```

```python
rf_cv.best_score_
```

```python
rf_best = RandomForestClassifier(max_depth=50,max_features='sqrt',max_samples=0.5,\
                                 min_samples_leaf=1,min_samples_split=0.001,n_estimators=50,random_state = 0)
```

```python
rf_best.fit(X_train,y_train)
```

```python
y_pred = rf_best.predict(X_test)
```

```python
print("Accuracy = ",accuracy_score(y_test, y_pred))
print("Precision = ",precision_score(y_test, y_pred))
print("Recall = ", recall_score(y_test, y_pred))
print("F1-score = ",f1_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred, labels = rf_best.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = rf_best.classes_ )
disp.plot()

```

```python
#feature_importance = rf_best.feature_importances_
#feature_names = X_train.columns
#feature_importance.plot.bar()
import matplotlib.pyplot as plt

importances = rf_best.feature_importances_

forest_importances = pd.Series(importances, index=X.columns)

fig, ax = plt.subplots()
forest_importances.plot.barh(ax=ax)
```
